# Read csv data function

```python
def read_csv(filename):
    df = pd.read_csv(filename).drop('Unnamed: 0', 1)
    df.name = filename.replace('.csv','')
    return df
```

# Visualization  Function

```python
def scatter_best_fit(df1, column1, df2, column2, selected_color):
    
    plt.scatter(x = df1[column1], y = df2[column2], color = selected_color)
    
    x1 = df1[column1]
    y1 = df2[column2]
    
    #m1,b1 = np.polyfit(x1,y1,1)
    #plt.plot(x1,m1*x1 + b1, '-')
    
    
    plt.xlabel(df1.name, fontsize=18)
    plt.ylabel(df2.name, fontsize=16)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
    #print(df1[column1].corr(df2[column2], method =  'pearson'))
```

```python
def plot_all_group(df1, df2):
    df1columns = list(df1.columns)[1:]
    df2columns = list(df2.columns)[1:]
    
    colors  = iter(cm.rainbow(np.linspace(0,1, len(df1columns))))
    
    for i in range(len(df1columns)):
        column1 = df1columns[i]
        column2 = df2columns[i]
        #scatter_best_fit(df1, column1, df2, column2, df3, column3)
        scatter_best_fit(df1, column1, df2, column2, next(colors))
    plt.show()
```

```python
def heat_map_df(income_nlz, children_nlz, Threashold = 0):
    
    income_levels = income_nlz.columns[1:]
    children_numbers = children_nlz.columns[2:]
    
    
    Children_Income_corr = pd.DataFrame(columns = children_numbers)
    Children_Income_corr.insert(0, "Income Level", income_levels)
    Children_Income_corr = Children_Income_corr.set_index(income_levels)
    
    #x-axis is number of children ever born
    #y-axis is income level
    for income_level in income_levels:
        for children_number in children_numbers:
            correlation = children_nlz[children_number].corr(income_nlz[income_level], method =  'pearson')
            if abs(correlation) <  Threashold:
                correlation = 0
            
            Children_Income_corr.at[income_level, children_number] = correlation

    Children_Income_corr = Children_Income_corr.drop("Income Level", axis = 1)  
    Children_Income_corr = Children_Income_corr[Children_Income_corr.columns].astype(float)
    return Children_Income_corr
```

# classify function

```python
def divide_according_to_mean(dataframe, grouping, title, drop_column = None):
    columns = dataframe.columns
    if drop_column != None:
        dataframe = dataframe.drop(columns[drop_column], axis =1 )
    cluster0 = pd.DataFrame(columns = dataframe.columns)
    cluster1 = pd.DataFrame(columns = dataframe.columns)
    cluster2 = pd.DataFrame(columns = dataframe.columns)
    cluster3 = pd.DataFrame(columns = dataframe.columns)
    for name in dataframe[title]:
        row = dataframe.loc[dataframe[title] == name]
        #print(row)
        if name in grouping[0]:
            cluster0 = cluster0.append(row, ignore_index = True)
        elif name in grouping[1]:
            cluster1 = cluster1.append(row, ignore_index = True)
        elif name in grouping[2]:
            cluster2 = cluster2.append(row, ignore_index = True)
        elif name in grouping[3]:
            cluster3 = cluster3.append(row, ignore_index = True)
    return cluster0, cluster1, cluster2, cluster3
```

# Step1 Grouping LGAs

```python
ChildrenVsIncome = read_csv("ChildrenVsIncome.csv")
```

```python
#normallize the data into [0,1]
multichildren_normalize = ChildrenVsIncome[ChildrenVsIncome.columns[1]] /  ChildrenVsIncome[ChildrenVsIncome.columns[-2]]
high_income_normalize = ChildrenVsIncome[ChildrenVsIncome.columns[2]]/  ChildrenVsIncome[ChildrenVsIncome.columns[-1]]
low_income_normalize = ChildrenVsIncome[ChildrenVsIncome.columns[3]]/  ChildrenVsIncome[ChildrenVsIncome.columns[-1]]

ChildrenVsIncome_normalize = pd.DataFrame()
ChildrenVsIncome_normalize.insert(0, "LGA", ChildrenVsIncome['LGA'])
ChildrenVsIncome_normalize.insert(1, "mother born >=3 children", multichildren_normalize)
ChildrenVsIncome_normalize.insert(2, "Income >= 1500/week", high_income_normalize)
ChildrenVsIncome_normalize.insert(3, "Income < 1500/week", low_income_normalize)
```

```python
colorset = np.arange(27)

mother_multichildren_ratio = ChildrenVsIncome[ChildrenVsIncome.columns[1]] /  ChildrenVsIncome[ChildrenVsIncome.columns[-2]]
female_highincome_ratio = ChildrenVsIncome[ChildrenVsIncome.columns[2]]/  ChildrenVsIncome[ChildrenVsIncome.columns[-1]]

#put scatter plot into 3 cluster
Children_HighIncome = []
for i in range(27):
    coordinate = [mother_multichildren_ratio[i],female_highincome_ratio[i]]
    Children_HighIncome.append(coordinate)
Children_HighIncome = np.array(Children_HighIncome)
kmeans = KMeans(n_clusters=4, random_state = 0).fit(Children_HighIncome)

xs = [x[0] for x in kmeans.cluster_centers_]
ys = [y[1] for y in kmeans.cluster_centers_]
print(kmeans.cluster_centers_)
```

Print out the grouping according to k-means algorith

```python
def nearest_centermean(x,y):
    centermeans = [list(mean) for mean in kmeans.cluster_centers_]
    smallest_distance = 1
    nearest_center_index = 0
    for centermean in centermeans:
        x0 = centermean[0]
        y0 = centermean[1]
        distance = math.hypot(x - x0, y - y0)
        if distance < smallest_distance:
            nearest_center_index = centermeans.index(centermean)
            smallest_distance = distance
    
    return nearest_center_index
    #this function return 0 1 2 3
    #represent which group should this data point be
```

```python

grouping = [[],[],[], []]
for i in range(27):
    x = multichildren_normalize[i]
    y = high_income_normalize[i]
    LGA = ChildrenVsIncome_normalize['LGA'][i]
    grouping[nearest_centermean(x,y)].append(LGA)
```

Based on this grouping, divide the original data set

```python
cluster0, cluster1, cluster2, cluster3 = divide_according_to_mean(ChildrenVsIncome_normalize, grouping, "LGA", -1)
```

```python
def normalize(df):
    #this function take a dataframe as input and used the copy of it to normalize the data
    columns = df.columns
    Total = columns[-1]
    NotStated = columns[-2]
    valid_total = df[Total] - df[NotStated]
    df_normalize = df.copy()
    for column in columns[1:-2]:
        df_normalize[column] = df[column] / valid_total
    df_normalize = df_normalize.drop([Total, NotStated], axis = 1)
    return df_normalize
    
```

# Step3 Welfare analysis on Each LGA group

```python
def normalize_welfare(df):
    columns = df.columns
    LGA_column = columns[0]
    df_normalized = df.copy()
    
    drop_list = []
    for i in range(0, len(columns)):
        if i % 2 == 0:
            drop_list.append(columns[i])
        else:
            df_normalized[columns[i]] = df_normalized[columns[i]] / df_normalized[columns[i + 1]]
    
    for drop in drop_list[1:]:
        df_normalized.drop(drop, axis = 1, inplace = True)

    return df_normalized.sort_values(LGA_column)
    
    
```

```python
welfare = read_csv("./Welfare data/welfare.csv")
```

```python
wel_cluster0, wel_cluster1, wel_cluster2, wel_cluster3= divide_according_to_mean(welfare, grouping, "LGA")
```

```python
def heat_map_df2(welfare_nlz, children_nlz, Threashold = 0):
    
    welfare_types = welfare_nlz.columns[1:]
    children_numbers = children_nlz.columns[2:]
    
    
    Children_Welfare_corr = pd.DataFrame(columns = children_numbers)
    Children_Welfare_corr.insert(0, "Welfare Type", welfare_types)
    Children_Welfare_corr = Children_Welfare_corr.set_index(welfare_types)
    
    #x-axis is number of children ever born
    #y-axis is income level
    for welfare_type in welfare_types:
        for children_number in children_numbers:
            correlation = children_nlz[children_number].corr(welfare_nlz[welfare_type], method =  'pearson')
            if abs(correlation) <  Threashold:
                correlation = 0
            
            Children_Welfare_corr.at[welfare_type, children_number] = correlation

    Children_Welfare_corr = Children_Welfare_corr.drop("Welfare Type", axis = 1)  
    Children_Welfare_corr = Children_Welfare_corr[Children_Welfare_corr.columns].astype(float)
    return Children_Welfare_corr
```
